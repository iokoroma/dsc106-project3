```python
import pandas as pd
import numpy as np
import altair as alt
import datetime as dt
import matplotlib.pyplot as plt
```

Context for categories: https://github.com/fivethirtyeight/data/tree/master/congress-resignations

1. more congressional democrats/republicans commit SA! (deceptive)
2. of congressional resignations up to 2018, the most happened in X decade.

```python
resigned = pd.read_csv('congressional_resignations.csv')
```

```python
resigned['Resignation Date'] = pd.to_datetime(resigned['Resignation Date'])
resigned['Year'] = [i.year for i in resigned['Resignation Date']]
# resigned['Year'].value_counts()
```

```python
sa = resigned.loc[(resigned['Category'] == 'X') | (resigned['Category'] == 'A')]
len(sa)
```

```python
sa_party = sa['Party'].value_counts()
color = ['red', 'blue']
bars = ('Republicans', 'Democrats')
x_pos = np.arange(len(bars))
fig = plt.figure(figsize=(10, 7))
plt.bar(sa_party.keys(), sa_party.values, edgecolor='k', color=color)
plt.xticks(x_pos, bars)
plt.xlabel('Political Party', weight='bold')
plt.ylabel('Resigned Members', weight='bold')
plt.title('Inappropriate Sexual Behavior among Congressional Republicans and Democrats')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
plt.hist(resigned['Year'], bins=118, edgecolor='k')
plt.xlabel('Year of Resignation', weight='bold')
plt.ylabel('Number of Resigned Members', weight='bold')
plt.title('Congressional Resignations by Year (1901-2018)')
plt.show()
```
